##Hash de una cadena

```python
import hashlib
def hasheo(m):
    encode= m.encode("utf-8")
    hash = hashlib.sha256(encode)

    return hash.hexdigest()

```

```python
print(hasheo("hola"))
```

##Generación de arbol de manera simple

```python
def arbol(x,y,z,w):
    x = hasheo(x)
    y = hasheo(y)
    z = hasheo(z)
    w = hasheo(w)

    h1=hasheo(str(x)+str(y))
    h2=hasheo(str(z)+str(w))

    return hasheo(str(h1)+str(h2))

```

```python
print(arbol("hola","Hola","chao","Chao"))
```

##Generación de merkle tree para entradas pares e impares

```python
def altarbol(x):
    print(x)
    if(len(x)==1):
        return x[0]
    y=[]
    if(len(x)%2==0):
        for i in range(0,len(x),+2):
            y.append(hasheo(x[i]+x[i+1]))
    else:
        for i in range(0,len(x)-1,+2):
            y.append(hasheo(x[i]+x[i+1]))
        y.append(x[len(x)-1])

    return altarbol(y)

def arbolp(x):
    y=[]
    for a in x:
       y.append(hasheo(a))
    return altarbol(y)
```

```python
x =["hola","Hola","chao","Chao"]
print(arbolp(x))
```

```python
x =["hola","Hola","chao","Chao","x","y","z"]
print(arbolp(x))
```
