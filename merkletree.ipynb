```python
import hashlib
def hasheo(m):
    encode= m.encode("utf-8")
    hash = hashlib.sha256(encode)

    return hash.hexdigest()

def altarbol(x):
    if(len(x)==1):
        return x[0]
    print(x)
    y=[]
    if(len(x)%2==0):
        for i in range(0,len(x),+2):
            y.append(hasheo(x[i]+x[i+1]))
    else:
        for i in range(0,len(x)-1,+2):
            y.append(hasheo(x[i]+x[i+1]))
        y.append(x[len(x)-1])

    return altarbol(y)

def arbolp(x):
    y=[]
    for a in x:
       y.append(hasheo(a))
    return altarbol(y)
```

```python
x =["hola","Hola","chao","Chao"]
print(arbolp(x))
```
